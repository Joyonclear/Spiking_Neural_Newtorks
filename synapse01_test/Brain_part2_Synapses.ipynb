```python
from brian2 import *
%matplotlib inline
```

```python
# config plot size
figure(figsize=(10,4))
start_scope()

# Each of which has the same differential equation.
eqs = '''
dv/dt = I/ms + (-v)/tau : 1
I : 1
tau : second
'''
# dv/dt = (I-v)/tau : 1

# But different values for parameters I and tau.
G = NeuronGroup(2, eqs, threshold='v>1', reset='v=0', method='euler')
G.I = [0.25, 0]
G.tau = [10, 40]*ms
# Input은 각각 0.2 0.1로 바로 들어간다
# tau값은 각각 다르게 설정 => 클수록 천천히 줄어든다

# Comment these two lines out to see what happens without Synapses
S = Synapses(G, G, on_pre='v_post += 0.2')
S.connect(i=0, j=1)
# 서로 각각(뉴런)의 출력을 입력으로 추가로 받는데 출력의 0.1만큼 받는다.

spikemon = SpikeMonitor(G)
M = StateMonitor(G, 'v', record=True)


run(50*ms)

plot(M.t/ms, M.v[0], label='Neuron 0')
plot(M.t/ms, M.v[1], label='Neuron 1')
# for t in spikemon.t:
#     axvline(t/ms, ls='--', c='r', lw=3)
    
xlabel('Time (ms)')
ylabel('v')
legend();
```

```python
# config plot size
figure(figsize=(20,8))
start_scope()

# Each of which has the same differential equation.
eqs = '''
dv/dt = I/ms + (-v)/tau : 1
I : 1
tau : second
'''
# dv/dt = (I-v)/tau : 1

# But different values for parameters I and tau.
G = NeuronGroup(2, eqs, threshold='v>1', reset='v=0', method='euler')
G.I = [0.2, 0]
G.tau = [10, 30]*ms
# Input은 각각 0.2 0.1로 바로 들어간다
# tau값은 각각 다르게 설정 => 클수록 천천히 줄어든다

# Comment these two lines out to see what happens without Synapses
S = Synapses(G, G, 'w : 1', on_pre='v_post += w')
S.connect(i=0, j=1)
S.w = 'j*0.2'
# 서로 각각(뉴런)의 출력을 입력으로 추가로 받는데 출력의 0.1만큼 받는다.

spikemon = SpikeMonitor(G)
M = StateMonitor(G, 'v', record=True)


run(50*ms)

plot(M.t/ms, M.v[0], label='Neuron 0')
plot(M.t/ms, M.v[1], label='Neuron 1')
# for t in spikemon.t:
#     axvline(t/ms, ls='--', c='r', lw=3)
    
xlabel('Time (ms)')
ylabel('v')
legend();
```
